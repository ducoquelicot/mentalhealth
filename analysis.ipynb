```python
x = df.klachten.to_list()
```

```python
import ast
```

```python
lst = []
for item in x:
    s = ast.literal_eval(item)
    lst.append(s)
```

```python
flat = [item for sublist in lst for item in sublist]
```

```python
import collections
```

```python
occ = collections.Counter(flat)
```

```python
occ.most_common()
```
